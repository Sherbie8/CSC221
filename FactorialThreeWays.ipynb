```python
# Compute Factorial using Three different programming techniques

# Definition: N! is the product of all numbers from 1 to N
# Example: 5! = 1 * 2 * 3 * 4 * 5
```

```python
# Method 1: List
def factorial1(n):
    #print(f'Computing factorial of {n}')

    # Make list of all ints 1 - N
    numbers = [x+1 for x in range(n)]
    #print(numbers)

    fact = 1
    for x in numbers:
        fact *= x
    return fact

print(factorial1(5))
print(factorial1(10))
print(factorial1(100))
```

```python
# Method 2: Traditional, don't make a list!
def factorial2(n):
    print(f'Computing factorial of {n}')

    fact = 1
    for x in range(n):
        #print(x)
        fact *= x+1
    return fact

print(factorial2(5))
print(factorial2(10))
print(factorial2(100))
```

```python
# Method 3: Recursive
def factorial3(n):
    print(f'Computing factorial of {n}')

    if(n == 1):
        return 1
    else:
        return n * factorial3(n-1)

print(factorial3(5))
print(factorial3(10))
print(factorial3(100))
```

```python
# Now we can use our factorial (any of them) to compute other things, like Combinations (see CSC 217!!)

def choose(n, r):
    '''
    How many sets of r objects can you make from a set of n things?
    '''
    return factorial1(n) / (factorial1(r) * factorial1(n - r))

help(choose)
print(choose(4,4))
print(choose(4,3))
print(choose(4,2))
print(choose(4,1))
print(choose(4,0))
```

```python
# Binomial Coefficients
# What are the coefficients we expect?
def coeffs(n):
    for i in range(n+1):
        print(choose(n,i))

coeffs(4)
coeffs(49)
coeffs(490)
```
