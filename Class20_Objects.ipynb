```python
# Class 20 11/7/2023
# Object Protocol
# Garbage Collection
# time.time()
#
class TestClass:
    #def __new__(cls, name, x):
    #    print(f'TestClass.__new__ creating a new {cls.__name__} object...')
    #    obj = object.__new__(cls)
    #    obj.name = name.capitalize()
    #    obj.x = x
    #    obj.big_list = [i**2 for i in range(10)]
    #    return obj
    def __init__(self, name, x):
        print(f'TestClass.__init__(self, {name}, {x})')
        self.name = name.capitalize()
        self.x = x
        self.big_list = [i**2 for i in range(10)]
    def __repr__(self):
        return f'TestClass({self.name}, {self.x})'
    def __str__(self):
        prefix = 'I am ' if self.name != 'Hodor' else ''
        return f'TestClass:\n  {prefix}{self.name}\n  {self.x}\n  {self.big_list[:10]})'
    def __del__(self):
        print(f'__del__() on {self.name=} {self.x=}')
```

```python
# With Timer
import time

class TestClass:
    #def __new__(cls, name, x):
    #    print(f'TestClass.__new__ creating a new {cls.__name__} object...')
    #    obj = object.__new__(cls)
    #    obj.name = name.capitalize()
    #    obj.x = x
    #    obj.big_list = [i**2 for i in range(10)]
    #    return obj
    def __init__(self, name, x):
        print(f'TestClass.__init__(self, {name}, {x})')
        start = time.time()
        self.name = name.capitalize()
        self.x = x
        self.big_list = [i**2 for i in range(1_000_000)]
        end = time.time()
        print(f'TestClass.__init__ took {(end - start) * 1000:0.00f} ms')
    def __repr__(self):
        return f'{__name__}({self.name}, {self.x})'
    def __str__(self):
        prefix = 'I am ' if self.name != 'Hodor' else ''
        return f'TestClass:\n  {prefix}{self.name}\n  {self.x}\n  {self.big_list[:10]})'
    def __del__(self):
        print(f'__del__() on {self.name=} {self.x=}')
```
